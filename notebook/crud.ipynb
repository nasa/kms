```python
import requests
```

## Demo of CRUD
### Create a concept

```python
headers = {
    'Content-Type': 'application/rdf+xml',
    'Accept': 'application/json'
}
response = requests.post(
    "https://cmr.sit.earthdata.nasa.gov/kms/concept",
    headers=headers,
    data=open('create.rdf', 'rb').read()
)
print(response.status_code)
print(response.text)
```

###  Retrieve a concept

```python
headers = {
    'Content-Type': 'application/rdf+xml',
    'Accept': 'application/json'
}
response = requests.get(
    "https://cmr.sit.earthdata.nasa.gov/kms/concept/071a00d6-29c6-460a-9d59-d593e430d263",
    headers=headers,
)
print(response.status_code)
print(response.text)
```

#### Update concept




```python
headers = {
    'Content-Type': 'application/rdf+xml',
    'Accept': 'application/json'
}
response = requests.put(
    "https://cmr.sit.earthdata.nasa.gov/kms/concept",
    headers=headers,
    data=open('update.rdf', 'rb').read()
)
print(response.status_code)
print(response.text)
```

#### Delete a concept


```python
headers = {
    'Content-Type': 'application/rdf+xml',
    'Accept': 'application/json'
}
response = requests.delete(
    "https://cmr.sit.earthdata.nasa.gov/kms/concept/071a00d6-29c6-460a-9d59-d593e430d263",
    headers=headers,
)
print(response.status_code)
print(response.text)
```

#### Retrieve all concepts

```python
headers = {
    'Content-Type': 'application/rdf+xml',
    'Accept': 'application/json'
}
response = requests.get(
    "https://cmr.sit.earthdata.nasa.gov/kms/concepts",
    headers=headers,
)
print(response.status_code)
print(response.text)
```
